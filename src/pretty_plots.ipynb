```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = '/home/kriti/ambivalent'
```

```python
# Set Seaborn style
sns.set_style("darkgrid")
sns.set_palette("tab10")  # Set a color palette

# Define parameters
thetas = [0.3, 0.3, 0.5, 0.5, 0.7, 0.7]
exp_bases = [6, 10, 6, 10, 6, 10]
labels = ['(0.25, 6)', '(0.25, 12)', '(0.5, 6)', '(0.5, 12)', '(0.7, 6)', '(0.7, 12)']
colors = sns.color_palette("tab10")

# Create subplots
fig, ax = plt.subplots(figsize=(7, 5))

# Plot the data
for i in range(len(thetas)):
    x = np.arange(0, 1, 0.1)
    y = 1 / (1 + np.exp(-exp_bases[i] * (x - thetas[i])))
    ax.plot(x, y, label=labels[i], color=colors[i])

# Set plot title and labels
ax.set_title('Global Trust Factor')
ax.set(xlabel='Training Time Completed %')
ax.legend()

# Save the plot
plt.savefig(PATH + '/images/global_trust.png',  bbox_inches='tight')

# Show the plot
plt.show()
```

```python
df1 = pd.read_csv(PATH + '/src/BiLSTM-train-labels-changed.csv')
df2 = pd.read_csv(PATH + '/src/BiLSTM-train-labels-corrected.csv')
columns = ["a", "b", "c", "d", "e", "f"]
labels = ['(0.25, 6)', '(0.25, 12)', '(0.5, 6)', '(0.5, 12)', '(0.7, 6)', '(0.7, 12)']

sns.set(style="darkgrid")
sns.color_palette("deep")
sns.set(font_scale=1)

plt.figure(figsize=(8, 4))
x_values = list(range(50))

train_df = pd.read_csv(PATH + '/train-predictions-BiLSTM-0.7-6.csv')
train_df = train_df[train_df['epoch'] == 0]
train_len = len(train_df)

for col, label in zip(columns, labels):
# Set Seaborn style
    sns.lineplot(x=x_values, y=df1[col]/train_len, label=label,  linestyle='-')

for col, label in zip(columns, labels):
# Set Seaborn style
    sns.lineplot(x=x_values, y=df2[col]/train_len, label=label,  linestyle='--')

# Set plot title and labels
plt.title("BiLSTM train labels flipped %")
plt.ylabel("")
plt.xlabel("Epoch")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.savefig(PATH + '/images/bilstm-train-labels-changed.png',  bbox_inches='tight')
```
